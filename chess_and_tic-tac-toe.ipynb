````python
# make board of chess
board_len = 400 # board length
height, width = board_len, board_len
board = np.zeros((height, width, 3), np.uint8)

# make board tiles
tile_l = 50
tile_color = (255, 255, 255) # white
for x in range(0, board_len, tile_l):
  for y in range(0, board_len, tile_l * 2):
    if (x // tile_l) % 2:
      y += tile_l
    board[x : x + tile_l, y : y + tile_l] = tile_color

# paint board
plt.axis("off")
plt.imshow(board)
plt.show()
````

```
2)Crie uma imagem 300 x 300 e utilizando o opencv com o python a transforme num jogo da velha com o X e O aonde dê empate

Sabendo:

cv2.line(imagem, ponto_inicial, ponto_final, cor, espessura)
cv2.rectangle(imagem, canto_superior_esquerdo, canto_inferior_direito, cor, espessura)
cv2.circle(imagem, ponto_centro, raio, cor, espessura)
Onde:
Ponto e canto é a coordenada do ponto(x,y)
Cor é (b, g, r)
```

````python
# make board of "tic-tac-toe" (Jogo da Velha)
board_len = 300
height, width = board_len, board_len
board = np.zeros((height, width, 3), np.uint8)

# make board lines
line_color = (255, 255, 255) # white
line_thickness = 10
# horizontal lines
cv2.line(board, (int(board_len * 1/3), 0), (int(board_len * 1/3), board_len), line_color, line_thickness)
cv2.line(board, (int(board_len * 2/3), 0), (int(board_len * 2/3), board_len), line_color, line_thickness)
# vertical lines
cv2.line(board, (0, int(board_len * 1/3)), (board_len, int(board_len * 1/3)), line_color, line_thickness)
cv2.line(board, (0, int(board_len * 2/3)), (board_len, int(board_len * 2/3)), line_color, line_thickness)


# MAKE X AND O MOVES:

x_color = (0, 0, 255) # blue
o_color = (255, 0, 0) # red
gap = 20

# X
for i, j in (1, 1), (0, 0), (1, 2), (0, 1), (2, 0):
  pos_x, pos_y = int(board_len * j/3), int(board_len * i/3)
  cv2.line(
    board, 
    (gap + pos_x, gap + pos_y), 
    (int(board_len * 1/3) - gap + pos_x, int(board_len * 1/3) - gap + pos_y), 
    x_color, 
    line_thickness
  )
  cv2.line(
    board, 
    (gap + pos_x, int(board_len * 1/3) - gap + pos_y), 
    (int(board_len * 1/3) - gap + pos_x, gap + pos_y), 
    x_color, 
    line_thickness
  )

# O
for i, j in (0, 2), (2, 2), (1, 0), (2, 1):
  pos_x, pos_y = int(board_len * j/3), int(board_len * i/3)
  cv2.circle(
    board, 
    (int(board_len * 1 / (3*2)) + pos_x, int(board_len * 1 / (3*2)) + pos_y),
    int(board_len * 1 / (3*2)) - gap, 
    o_color, 
    line_thickness
  )


# paint board
plt.axis("off")
plt.imshow(board)
plt.show()

# result
print("Empate!")
````
